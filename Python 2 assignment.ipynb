6-What is a string in Python, and how is it different from a list?

ans- A string in Python is an immutable sequence of characters. It's used to represent text.

Here's how strings differ from lists:

*   **Mutability:** Strings are immutable, meaning you cannot change individual characters within an existing string. Lists are mutable, allowing you to modify, add, or remove elements.
*   **Data Type:** Strings specifically store sequences of characters. Lists can store elements of any data type (integers, floats, strings, other lists, etc.).
*   **Syntax:** Strings are typically enclosed in single, double, or triple quotes (`'hello'`, `"world"`, `"""multiline"""`). Lists are enclosed in square brackets (`[1, 'a', True]`).
*   **Use Cases:** Strings are used for text manipulation, representing names, sentences, etc. Lists are used for ordered collections of items, where the order and ability to modify the collection are important.

8-What is a hash table, and how does it relate to dictionaries in Python?

 ans-A hash table is a data structure that stores key–value pairs and allows very fast data retrieval using a key.

 Relation to Python Dictionaries

In Python, a dict is implemented as a hash table.

Keys in a dictionary are hashed to find where the corresponding value is stored.

Because of hashing:

Lookup, insertion, and deletion in a dictionary are typically O(1) (constant time).

Keys must be hashable (immutable types like str, int, tuple of immutables).

17-Describe the time complexity of accessing elements in a dictionary?

ans-Accessing elements in a Python dictionary is generally O(1) — constant time — thanks to its hash table implementation.

How it works

Hashing the key → Python applies a hash function to the key to find an index in the internal table.

Direct access → Jumps directly to the memory location for that key’s value.

Collision handling → If multiple keys hash to the same index, Python checks them sequentially (usually very few).

Time Complexity
Operation                      	Average Case               	Worst Case*
Access value by key (dict[k])     	O(1)                     	O(n)
Insert key-value pair              	O(1)                    	O(n)
Delete key-value pair	              O(1)                     	O(n)


18-In what situations are lists preferred over dictionaries?

ans- Lists are preferred over dictionaries when you don’t need key-value mapping and order or sequence is more important than named access.

Situations Where Lists Are Better
1. Ordered Collections

Lists maintain a defined order and allow access by position (index), which is useful when order matters.

2. Allow Duplicates

Lists can store the same value multiple times, unlike dictionary keys which must be unique.

3. Simple Collections of Values

When you just need a group of related values without associating them with specific keys.

4. Frequent Iteration

Lists are efficient for iterating over all items when you don’t need to map them to identifiers.

5. Index-based Access

When the position of an item is meaningful (e.g., leaderboard rankings, queue systems).

6. Memory Efficiency

Lists can use slightly less memory than dictionaries for small datasets since dictionaries store both keys and values along with hash overhead.

19-Why are dictionaries considered unordered, and how does that affect data retrieval|?

ans-Dictionaries are considered unordered because, historically in Python, they did not guarantee that the order of key-value pairs in memory matched the order in which they were inserted.

Why Dictionaries Were Unordered

Dictionaries use a hash table internally.

Keys are placed in memory based on their hash value, not their insertion order.

This means the order of items could vary depending on:

Python version

Insertions and deletions

Hash collisions

Effect on Data Retrieval

Before Python 3.7:
You could not rely on the order of items when iterating over a dictionary — the sequence could appear random.

From Python 3.7 onward:
Python preserves insertion order as an implementation detail (and officially guarantees it from Python 3.7+).
However, even now:

Dictionaries are still conceptually unordered — meaning order should not be your main reason to choose them.

Retrieval by key is based on hashing, not position.
